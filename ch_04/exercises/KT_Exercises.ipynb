## Q1a - Alternative

```python
# So I'll reverse engineer the missing datasets
for ticker in ['FB', 'AAPL', 'AMZN', 'NFLX', 'GOOG']:
    filt = (df['ticker'] == ticker)
    out_df = df.loc[filt].copy()
    out_df.drop(columns = ['ticker'], inplace=True)
    out_df.to_csv(f'KTs-{ticker}.csv', index=False)
    # Debugging
    print(f'\n\n\n\n\n{ticker=}')
    print(f'{filt=}')
    print(f'{out_df}')
```

```python
# And then do it again
main_df = pd.DataFrame()
for ticker in ['FB', 'AAPL', 'AMZN', 'NFLX', 'GOOG']:
    temp_df = pd.read_csv(f'KTs-{ticker}.csv')
    temp_df['ticker'] = ticker
    main_df = pd.concat([main_df, temp_df])
main_df
```
