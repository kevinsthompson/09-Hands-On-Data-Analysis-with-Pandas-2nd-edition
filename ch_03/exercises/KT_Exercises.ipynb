## Q6 - All in One (actually 2) 

```python
df = pd.read_csv(
    'covid19_cases.csv').\
        assign(date = lambda x: pd.to_datetime(x['dateRep']))\
        .set_index('date')\
        .sort_index()\
        .replace(
            {'United_States_of_America': 'USA',
             'United_Kingdom': 'UK'})

# df.dtypes
# df
```

```python
filt = (
    df['countriesAndTerritories'].isin(
        ['Argentina','Brazil','Chile','Colombia',
         'India','Italy','Mexico','Peru','Russia',
         'Spain','Turkey','UK','USA']))

df[filt].reset_index().pivot(
    index='date', 
    columns='countriesAndTerritories', 
    values='cases'
    ).fillna(0)


# df[
#     df.countriesAndTerritories.isin([
#         'Argentina', 'Brazil', 'China', 'Colombia', 'India', 'Italy', 
#         'Mexico', 'Peru', 'Russia', 'Spain', 'Turkey', 'UK', 'USA'
#     ])]

# df.drop(index=df[~filt].index, inplace=True)
# df
# covid[
#     covid.countriesAndTerritories.isin([
#         'Argentina', 'Brazil', 'China', 'Colombia', 'India', 'Italy', 
#         'Mexico', 'Peru', 'Russia', 'Spain', 'Turkey', 'UK', 'USA'
#     ])
# ].reset_index().pivot(index='date', columns='countriesAndTerritories', values='cases').fillna(0)
```
